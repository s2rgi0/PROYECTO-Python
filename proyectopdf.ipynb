```python
from fpdf import FPDF

```

```python
pdf = FPDF()


pdf.add_page()
pdf.set_font("Arial")

pdf.image("Template.png", 0,0)

pdf.text(10,10, "Probando el pdf")
pdf.output("Presupuesto.pdf")


```
